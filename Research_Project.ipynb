```python
from typing_extensions import final
#!wget -q https://archive.ics.uci.edu/ml/machine-learning-databases/00275/Bike-Sharing-Dataset.zip
#!unzip -q Bike-Sharing-Dataset.zip
#!pip install pycaret[full]
import pycaret
from pycaret.regression import *
import pandas as pd


df = pd.read_csv('hour.csv')

s = setup(df, target = 'cnt', session_id = 123, categorical_features= ['season', 'mnth' , 'hr'], ignore_features= ['instant', 'dteday', 'casual', 'registered'] )
best = compare_models()
tuned_best = tune_model(best)
final_best = finalize_model(tuned_best)
holdout_pred = predict_model(final_best, data = df)
df['Predictions'] = holdout_pred['prediction_label']
df.to_csv('predictions.csv', index=False)
save_model(holdout_pred, 'best_regression_model')
```
